```python
#Create a bar chart of average salary by title.
avg_salarydf = pd.read_sql_query("SELECT ROUND(AVG(salaries.salary), 2), titles.title FROM salaries LEFT JOIN titles ON titles.emp_no=salaries.emp_no GROUP BY titles.title;", connection)
avg_salarydf
```

```python
colors = ['yellow', 'dodgerblue', 'navy', 'red', 'darkorange',
         'forestgreen', 'skyblue']

bar_chart = avg_salarydf.plot.bar(y='round', x='title', colors=colors, legend=False, rot=0, figsize=(9,5))

plt.xlabel("Titles")
plt.ylabel("Salaries")
plt.title("Average Salary by Title")

plt.savefig("C:\\Users\\richn\\Desktop\\avg_salary.png")

plt.show()
plt.tight_layout()
```
